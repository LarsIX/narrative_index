```python
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
# find root 
root = Path.cwd().parent

# build path to VIX data
vix_path = root / "data" / "external" / "VIX_History.csv"

# load data
df = pd.read_csv(vix_path)
```

```python
# convert to datetime
df["date"] = pd.to_datetime(df["DATE"], format="%m/%d/%Y")

# Convert as string
df["date"] = df["date"].dt.strftime("%Y-%m-%d")

# verify
df["date"].head()
```

```python
# drop all but close
drop_col = ["OPEN","HIGH","LOW","DATE"]

# drop cols
df_sub = df.drop(drop_col,axis=1)

# subset for April 2023 - July 2025
df_sub = df_sub[(df_sub["date"] >= "2023-04-01") & (df_sub["date"] < "2025-06-16")] 
```

```python
# calc log growth
df_sub["log_growth_closed"] = np.log(df_sub["CLOSE"]) - np.log(df_sub["CLOSE"].shift(1))
df_sub
```

```python
# define output path
out_path = root / "data" / "processed" / "variables" / "log_growth_VIX.csv"

# write to out_path
df_sub.to_csv(out_path, index=False)
out_path
```
